```python
def GetListOfModel():
    # Find the element
    element = driver.find_element(By.XPATH, '/html/body/snapon-app/snapon-epc/div/split-panel/split-area[1]/part-search/div/div/global-search/div[2]/div[1]/ag-grid-angular/div/div[2]/div[2]/div[3]/div[2]/div/div')
    element_html = element.get_attribute('outerHTML')
    # Parse the HTML with BeautifulSoup
    soup = bs4.BeautifulSoup(element_html, 'html.parser')

    itemlist = soup.findAll('div',{'class':'ag-cell ag-cell-not-inline-editing ag-cell-auto-height ag-cell-value ag-cell-wrap-text'})
    i=0
    step=0.5
    itemModelList = []
    while i < 1:        
        #Scroll Down in step
        left_menu = driver.find_element(By.XPATH, "/html/body/snapon-app/snapon-epc/div/split-panel/split-area[1]/part-search/div/div/global-search/div[2]/div[1]/ag-grid-angular/div/div[2]/div[2]/div[3]")
        # Scroll the left menu to 50% of its height using JavaScript
        scroll_script = "arguments[0].scrollTop = arguments[0].scrollHeight * "+ str(i) +";"
        driver.execute_script(scroll_script, left_menu)

        #Get List
        element = driver.find_element(By.XPATH, '/html/body/snapon-app/snapon-epc/div/split-panel/split-area[1]/part-search/div/div/global-search/div[2]/div[1]/ag-grid-angular/div/div[2]/div[2]/div[3]')
        element_html = element.get_attribute('outerHTML')
        soup = bs4.BeautifulSoup(element_html, 'html.parser')
        itemlist = soup.findAll('div',{'class':'ag-cell ag-cell-not-inline-editing ag-cell-auto-height ag-cell-value ag-cell-wrap-text'})
        itemlist = [ x.text for x in itemlist]
        itemModelList.extend(itemlist)
        # Increment i by the step value
        i += step

    itemModelList = set(itemModelList)
    itemModelList = {item for item in itemModelList if any(char.isdigit() for char in item.split('-')[0]) and len(item.split('-')[0].replace(".", "").replace(" ", "")) in range(3, 7)}
    return(len(itemModelList))
```

```python
def NumberofSearchResult():    
    SearchCountXPATH = (By.XPATH,"/html/body/snapon-app/snapon-epc/div/split-panel/split-area[1]/part-search/div/div/global-search/div[1]/div") #when found                         
    try:
        WebDriverWait(driver, 3).until(EC.visibility_of_element_located(SearchCountXPATH))
    except TimeoutException:
        try :
            SearchCountXPATH = (By.XPATH,"/html/body/snapon-app/snapon-epc/div/split-panel/split-area[1]/part-search/div/div/div[2]/div[1]/div")   #when not found 
            WebDriverWait(driver, 3).until(EC.visibility_of_element_located(SearchCountXPATH))
        except TimeoutException:
            pass
        else:
            return(0)
    else:
        pass   
    
    element = driver.find_element(By.XPATH,'/html/body/snapon-app/snapon-epc/div/split-panel/split-area[1]/part-search/div/div/global-search/div[1]/div')
    element_html = element.get_attribute('outerHTML')
    soup = bs4.BeautifulSoup(element_html, 'html.parser')
    numResult = soup.text
    numResult = int(numResult.split(':')[1].replace(' ',''))
    return(numResult)
```

```python
def GetGroupMapping():
    try:
        element = driver.find_element(By.XPATH,'/html/body/snapon-app/snapon-epc/div/split-panel/split-area[1]/epc-parts/div/split-panel/split-area[3]/div/parts-filter/div/select-button/div')
    except NoSuchElementException:
        mapType = {'A':''}
        return(mapType)
    
    element_html = element.get_attribute('outerHTML')
    soup = bs4.BeautifulSoup(element_html, 'html.parser')
    soup = soup.findAll('div',{'class':'ui-button ui-widget ui-state-default ui-button-text-only ng-star-inserted'})
    

    if soup[0].text.find(':')>=0:
        data = [x.text.split(' - ') for x in soup ]    
        # Initialize an empty dictionary
        result_dict = {}
        # Iterate over each sublist in the data list
        for sublist in data:
            # Extract the key and value from each sublist
            key, value = sublist[0].split(': ')
            # Add the key-value pair to the dictionary
            result_dict[key] = value

        mapType = result_dict
    else:
        mapType = dict([x.text.split(' - ') for x in soup ])
    return(mapType)
```

```python
def getListofSubTable():
    element = driver.find_element(By.XPATH,'/html/body/snapon-app/snapon-epc/div/split-panel/split-area[1]/epc-parts/div/split-panel/split-area[3]/parts-list/div/ag-grid-angular/div/div[2]/div[2]/div[3]/div[2]/div/div')
    element_html = element.get_attribute('outerHTML')
    soup = bs4.BeautifulSoup(element_html, 'html.parser')
    soup = soup.findAll('div',{'aria-label':'Press SPACE to deselect this row.'})
    return(soup)
```

```python
def ScrollThrough():
    PartList=[]
    i=0
    step=0.1
    while i < 1:        
        #Scroll Down in step
        centerList = driver.find_element(By.XPATH, "/html/body/snapon-app/snapon-epc/div/split-panel/split-area[1]/epc-parts/div/split-panel/split-area[3]/parts-list/div/ag-grid-angular/div/div[2]/div[2]/div[3]")
        # Scroll the left menu to 50% of its height using JavaScript
        scroll_script = "arguments[0].scrollTop = arguments[0].scrollHeight * "+ str(i) +";"
        driver.execute_script(scroll_script, centerList)        
        PartList.extend(getListofSubTable())
        i += step
        time.sleep(0.5)
    return(PartList)
    
```

```python
def getTableNew(result):
    ResultTable = pd.DataFrame()
    for x in list(set(result))[:]:
        PartNo = ''
        AppliedDate = ''
        position = ''
        Quantity = ''
        A = ''
        B = ''
        C = ''
        D = ''
        E = ''
        F = ''
        G = ''
        H = ''
        I = ''
        Remarks = ''
        try:
            PartNo = x.findAll('div', {'col-id': 'formattedPartNumber'})[0].text
            AppliedDate = x.findAll('div', {'col-id': 'APPLIED_DATE'})[0].text    
            position = x.findAll('div', {'col-id': 'position'})[0].text  
            Quantity = x.findAll('div', {'col-id': 'quantity'})[0].text  
            A = x.findAll('div', {'col-id': 'A_U'})[0].text
            B = x.findAll('div', {'col-id': 'B_U'})[0].text
            C = x.findAll('div', {'col-id': 'C_U'})[0].text
            D = x.findAll('div', {'col-id': 'D_U'})[0].text
            E = x.findAll('div', {'col-id': 'E_U'})[0].text
            F = x.findAll('div', {'col-id': 'F_U'})[0].text
            G = x.findAll('div', {'col-id': 'G_U'})[0].text
            H = x.findAll('div', {'col-id': 'H_U'})[0].text
            I = x.findAll('div', {'col-id': 'I_U'})[0].text
            Remarks = x.findAll('div', {'col-id': 'remarks'})[0].text
        except IndexError:
            pass
        tempTable = pd.DataFrame(columns=['PartNo', 'AppliedDate', 'position', 'Quantity','A','B','C','D','E','F','G','H','I','Remarks'])
        tempTable.loc[0] = [PartNo,AppliedDate,position, Quantity,A,B,C,D,E,F,G,H,I,Remarks]
        ResultTable = pd.concat([tempTable, ResultTable], ignore_index=True)
    ResultTable = ResultTable.drop_duplicates()
    return ResultTable
```

```python
def getDetail(PartNo):    
    figureClick = (By.XPATH,'/html/body/snapon-app/snapon-epc/div/split-panel/split-area[1]/part-search/div/div/global-search/div[2]/div[2]/ag-grid-angular/div/div[2]/div[2]/div[3]/div[2]/div/div/div/div[2]')
    WebDriverWait(driver, 10).until(EC.visibility_of_element_located(figureClick))

    element = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '/html/body/snapon-app/snapon-epc/div/split-panel/split-area[1]/part-search/div/div/global-search/div[2]/div[2]/ag-grid-angular/div/div[2]/div[2]/div[3]/div[2]/div/div/div/div[2]')))    
    element = driver.find_element(By.XPATH,'/html/body/snapon-app/snapon-epc/div/split-panel/split-area[1]/part-search/div/div/global-search/div[2]/div[2]/ag-grid-angular/div/div[2]/div[2]/div[3]/div[2]/div/div/div/div[2]')
    element.click()

    detailTableVisible = (By.XPATH,'/html/body/snapon-app/snapon-epc/div/split-panel/split-area[1]/epc-parts/div/split-panel/split-area[3]/parts-list/div/ag-grid-angular/div/div[2]/div[2]')
    WebDriverWait(driver, 10).until(EC.visibility_of_element_located(detailTableVisible))

    waitingSign = (By.XPATH,'/html/body/div[2]/div/div')
    WebDriverWait(driver, 120).until(EC.invisibility_of_element_located(waitingSign))

    result = ScrollThrough()
    result = getTableNew(result)
    mapping = GetGroupMapping()
    mapping = pd.DataFrame(mapping, index=[0])
    mapping['PartNo']=PartNo
    return(result,mapping)
    
```

```python
def closeDetail():
    element = driver.find_element(By.XPATH,'/html/body/snapon-app/snapon-epc/div/div[1]/div[3]/epc-breadcrumb/div/ul/li[1]/a/img')
    element.click()

    #Waiting for page is ready.
    PageReady = (By.XPATH,'/html/body/snapon-app/snapon-epc/div/split-panel/split-area[1]/epc-navigation/div/div[2]/div/div/div/ag-grid-angular/div/div[2]')
    WebDriverWait(driver, 60).until(EC.visibility_of_element_located(PageReady))
```

```python
def getImage(PartNo):
    element= driver.find_element(By.XPATH,'/html/body/snapon-app/snapon-epc/div/split-panel/split-area[1]/epc-parts/div/split-panel/split-area[2]/div[2]/parts-image/div/div/div/div[2]/img[3]')
    element.click()

    ImageReady = (By.XPATH,'/html/body/snapon-app/snapon-epc/div/split-panel/split-area[1]/epc-parts/div/div/div[2]/parts-image/div/snapon-imageviewer-component/table/tr[1]/td/canvas[3]')
    WebDriverWait(driver, 10).until(EC.visibility_of_element_located(ImageReady))

    img_element = driver.find_element(By.XPATH, '/html/body/snapon-app/snapon-epc/div/split-panel/split-area[1]/epc-parts/div/div/div[2]/parts-image/div/snapon-imageviewer-component/table/tr[1]/td/canvas[3]')
    location = img_element.location
    size = img_element.size

    try:
        driver.save_screenshot('page_screenshot.png')
    except TimeoutException:
        driver.save_screenshot('page_screenshot.png')
    finally:
        pass

    # Use PIL to crop the image
    x = location['x']
    y = location['y']
    width = size['width']
    height = size['height']
    im = Image.open('page_screenshot.png')
    im = im.crop((int(x), int(y), int(x + width), int(y + height)))
    im.save('images/ISU-'+str(PartNo)+'.png')

    element = driver.find_element(By.XPATH,'/html/body/snapon-app/snapon-epc/div/split-panel/split-area[1]/epc-parts/div/div/div[2]/parts-image/div/div/div/div[2]/img[3]')
    element.click()

    homeReady = (By.XPATH,'/html/body/snapon-app/snapon-epc/div/div[1]/div[3]/epc-breadcrumb/div/ul/li[1]/a/img')
    WebDriverWait(driver, 30).until(EC.visibility_of_element_located(homeReady))
    
```
